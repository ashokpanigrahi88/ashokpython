```python
import pandas as pd
import numpy as np
```

```python
url = ('https://raw.github.com/pandas-dev/pandas/master/pandas/tests/data/tips.csv')
url
tips = pd.read_csv(url)
tips.shape
```

```python
#select total_bill, tip, smoker, timw from tips limit 5
tips[['total_bill','tip','smoker','time']].head()
```

```python
#select * from tips where time = 'Dinner' limit 5
tips[tips['time'] == 'Dinner'].describe()
```

```python
is_dinner = tips['time'] == 'Dinner'
is_dinner.value_counts()
tips['total_bill'].max()
```

```python
#select * from tips where time = 'Dinner' and tip > 5
tips[(tips['time'] == 'Dinner' ) & (tips['tip'] < 5)]
```

```python
# select * From Tips where size >= 5 OR Total_bill > 45
tips[(tips['size'] <5) & (tips['total_bill'] > 45)]
```

```python
frame = pd.DataFrame({'col1': ['A','B',np.NaN,'C'],
                    'col2':['F',np.NaN,'G','H']})
frame
```

```python
# Select * from Frame where col2 is null
frame[(frame['col2'].isna()) | (frame['col1'].isna())]
frame[(frame['col2'].notna()) & (frame['col1'].notna())]
```

```python
#select sex, count(1) from tips group by sex:
tips.groupby('sex').size()
tips.groupby('sex').count()
tips.groupby(['sex','time']).sum()
```

```python
#select day, AVG(tip), count(1) from tips Group by day
tips.groupby('day').agg({'tip': np.mean, 'day': np.size})
```

```python
#select smoker, day, count(1), avg(tip) from tips group by smoker , day 
tips.groupby(['smoker','day']).agg({'tip': [np.size,np.mean] })
```
